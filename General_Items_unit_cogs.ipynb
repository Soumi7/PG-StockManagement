```
import pandas as pd
df=pd.read_csv("drive/My Drive/MERGED_FINAL_DATASET.csv")
```

```
df
```

```
df=df.drop(['Unnamed: 0',	'date', 'ishol/week',	'quantity','monthly_Avgtemp',	'monthly_avg_FeelsLikeC',	'monthly_avg_HeatIndexC',	'monthly_avg_cloudcover',	'monthly_avg_humidity'],axis=1)
```

So we create a dictionary to store the separated datasets with key as name of each group

```
d = {}

for frame, data in df.groupby('group'):
    d[frame] = data
```

```
print(d)
```

```
for key in d.keys():
  d[key]=d[key].drop(['group'],axis=1)
  
```
