```python
# function to categorise the heat into 3 different levels ... 0=low( <= -1 ) , 1= normal ( >-1 , <=14 ), 2= high ( >14,<=27 )

def cat_heat(data):
    monthly_heatindex=[]
    for heatindex in data["monthly_avg_HeatIndexC"]:
        if heatindex < -2:
            monthly_heatindex.append(0)
        elif heatindex >=-1 and heatindex<=14:
            monthly_heatindex.append(1)
        else:
            monthly_heatindex.append(2)
    return monthly_heatindex
            
```

```python
heat_index=cat_heat(weather_data)
```

```python
weather_data["monthly_avg_HeatIndexC"]=heat_index
```

```python

cloudcover=df_copy["cloudcover"].unique()
cloudcover.sort()
len(cloudcover)
```

```python
# function to categorise the heat into 3 different levels ... 0=clear sky( < 250 ) , 1= scattered clouds
# ( >25 , <=50 ), 2= broken clouds ( >50,<=75 )  3= overcast/ cloudy (>75)


def cat_cloud(data):
    monthly_cloudcover=[]
    for cloudcover in data["monthly_avg_cloudcover"]:
        if cloudcover < 25:
            monthly_cloudcover.append(0)
        elif cloudcover >=25 and cloudcover<50:
            monthly_cloudcover.append(1)
        elif cloudcover >=50 and cloudcover<75:
            monthly_cloudcover.append(2)    
        else:
            monthly_cloudcover.append(3)
    return monthly_cloudcover
            
```

```python
weather_data["monthly_avg_cloudcover"]=cat_cloud(weather_data)
```
